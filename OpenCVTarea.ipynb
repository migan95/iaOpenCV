```python
#Corregimos cambios anteriores volviendo a setear la imagen original
imagen_corregida=cv2.cvtColor(img,cv2.COLOR_RGB2BGR)
plt.imshow(imagen_corregida)
```

```python
imagen_volteada = cv2.flip(imagen_corregida,0)
```

```python
plt.imshow(imagen_volteada)
```

```python
#Dibujar circulo en la posicin especificada (el reloj)
cv2.circle(imagen_volteada,center=(350,400),radius=50,color=(255,0,0),thickness=5)
```

```python
plt.imshow(imagen_volteada)
```

```python
#Dibujar triangulo, se compone de 3 lineas para simular un triangulo
cv2.line(imagen_volteada,(550,400),(500,475),(255,0,0),5)
cv2.line(imagen_volteada,(500,475),(600,475),(255,0,0),5)
cv2.line(imagen_volteada,(600,475),(550,400),(255,0,0),5)
```

```python
plt.imshow(imagen_volteada)
```
